```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("creditcard.csv")
print("Dataset Shape:", df.shape)
print(df.head())
corr = df.corr(numeric_only=True)
target_corr = corr["Class"].sort_values(ascending=False)
print("\n Correlation of all features with Class:")
print(target_corr)
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap – Credit Card Fraud Dataset")
plt.show()
top_features = target_corr[1:4]
print("\n Top 3 Features Most Related to Fraud (Class):")
print(top_features)
print("\n Insights:")
print("1 The features with highest correlation to fraud (Class) are often V17, V14, and V12.")
print("2 These components indicate strong patterns that help the model distinguish fraud from normal transactions.")
print("3 Amount has very weak correlation, showing that fraud isn’t necessarily about transaction size.")

```
