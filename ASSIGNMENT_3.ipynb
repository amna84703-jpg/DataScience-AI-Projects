```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("creditcard.csv")
# ---------------------------
# 1. Bar Chart: Class Imbalance
# ---------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Class", hue="Class", data=df, palette="Set2", legend=False)
plt.title("Class Distribution (Fraud vs Non-Fraud)")
plt.show()
# ---------------------------
# 2. Histogram: Transaction Amounts
# ---------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["Amount"], bins=50, kde=True, color="skyblue")
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.show()
# ---------------------------
# 3. Boxplot: Amount by Class
# ---------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="Class", y="Amount", hue="Class", data=df, palette="Set3", legend=False)
plt.title("Transaction Amounts by Class")
plt.show()
# ---------------------------
# 4. Correlation Heatmap (selected features)
# ---------------------------
plt.figure(figsize=(10,6))
corr = df[["V14", "V17", "V12", "V10", "Amount", "Class"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Key Features with Fraud Class")
plt.show()
# ---------------------------
# 5. Scatter Plot: Time vs Amount
# ---------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x="Time", y="Amount", hue="Class", data=df, alpha=0.6, palette="Set1")
plt.title("Transaction Time vs Amount (Fraud Highlighted)")
plt.show()

```
