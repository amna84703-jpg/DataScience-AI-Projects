```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
df = pd.read_csv("/content/creditcard.csv")
df_small = df.sample(200000, random_state=42).copy()
import random
transaction_types = [
    "online shopping payment",
    "atm cash withdrawal",
    "international purchase",
    "mobile recharge",
    "utility bill payment",
    "restaurant dining",
    "local store shopping",
    "fuel pump payment"
]
df_small["transaction_description"] = [random.choice(transaction_types) for _ in range(len(df_small))]
df_small.head()
stop_words = set(stopwords.words("english"))
def clean_text(text):
    tokens = word_tokenize(text.lower())
    cleaned = [word for word in tokens if word.isalpha() and word not in stop_words]
    return " ".join(cleaned)
df_small["cleaned_text"] = df_small["transaction_description"].apply(clean_text)
df_small[["transaction_description", "cleaned_text"]].head()
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_small["cleaned_text"])
print("TF-IDF Shape:", tfidf_matrix.shape)
from scipy.sparse import hstack
numeric_features = df_small.drop(["transaction_description", "cleaned_text", "Class"], axis=1)
combined_features = hstack([tfidf_matrix, numeric_features])
print("Final Feature Matrix Shape:", combined_features.shape)
X = combined_features
y = df_small["Class"]
print("NLP pipeline completed")
```
